```python
import numpy as np
import  pandas as pd
```

```python
cars = pd.read_csv('CarsC2.csv')
cars
```

```python
cars = cars[cars[' Box_State'].notna()]
cars = cars[cars['Name'] != "Sepeda Facific Invert3 Variant 1 Variant"]
cars
```

```python
cars
```

```python
cars = cars[['X','Z']]
cars
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example: Load your data into a pandas DataFrame
# data = pd.read_csv('path_to_your_csv_file.csv')

# For illustration, let's assume your DataFrame looks like this:
# data = pd.DataFrame({
#     'x': [x1, x2, x3, ...],
#     'y': [y1, y2, y3, ...]
# })

# Generate a 2D histogram
heatmap, xedges, yedges = np.histogram2d(cars['X'], cars['Z'], bins=(50,50))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Plot the heatmap
plt.clf()
plt.imshow(heatmap, extent=extent, origin='lower', cmap='hot', interpolation='nearest')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Assuming 'data' is your DataFrame with 'x' and 'y' columns

# Generate a 2D histogram with fewer bins for larger areas
heatmap, xedges, yedges = np.histogram2d(cars['X'], cars['Z'], bins=(128,128))  # fewer bins
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Apply a Gaussian filter for smoothing
heatmap_smoothed = gaussian_filter(heatmap, sigma=0.7)  # Adjust the sigma for more or less smoothing

# Plot the smoothed heatmap
plt.clf()
plt.imshow(heatmap_smoothed.T, extent=extent, origin='lower', cmap='hot', interpolation='bilinear')  # Use bilinear interpolation
plt.xlabel("X data from unity")
plt.ylabel("Z data from unity")
plt.show()
```

```python
import seaborn as sns

# Use seaborn to create a KDE plot
sns.kdeplot(cars['X'], cars['Z'], cmap="hot", shade=True, bw_adjust=0.5)  # Adjust bw_adjust for smoothing
plt.show()
```

# Combination

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Load your first dataset
# data1 = pd.DataFrame({'x': [...], 'y': [...]})

# Load your second dataset
# data2 = pd.DataFrame({'x': [...], 'y': [...]})

# Generate 2D histograms
heatmap1, xedges1, yedges1 = np.histogram2d(cars['X'], cars['Z'], bins=(64,64))
heatmap2, xedges2, yedges2 = np.histogram2d(humans['X'], humans['Z'], bins=(64,64))

# Apply Gaussian filter for smoothing
sigma = 0.8
heatmap1_smoothed = gaussian_filter(heatmap1, sigma=sigma)
heatmap2_smoothed = gaussian_filter(heatmap2, sigma=sigma)

# Define the extent of the heatmaps
extent1 = [xedges1[0], xedges1[-1], yedges1[0], yedges1[-1]]
extent2 = [xedges2[0], xedges2[-1], yedges2[0], yedges2[-1]]

# Plot the first heatmap
plt.imshow(heatmap1_smoothed.T, extent=extent1, origin='lower', cmap='Reds', alpha=0.5, interpolation='bilinear')

# Plot the second heatmap overlay with different color and transparency
plt.imshow(heatmap2_smoothed.T, extent=extent2, origin='lower', cmap='Blues', alpha=0.5, interpolation='bilinear')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming data1 and data2 are your DataFrames with 'x' and 'y' columns

# Plot KDE for the first dataset with one color
sns.kdeplot(cars['X'], cars ['Z'], cmap="hot", shade=True, bw_adjust=0.5, alpha=0.5)

# Plot KDE for the second dataset with another color on top
sns.kdeplot(humans['X'], humans['Z'], cmap="PuBu", shade=True, bw_adjust=0.5, alpha=0.5)

# Show the plot
plt.show()
```
